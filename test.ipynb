```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from PIL import Image
import cv2 as cv

# Paths to the models and test image
model1_path = "v3p2.h5"
model2_path = "vgg16M.h5"
model3_path = "DenseNetV2M.h5"
model4_path = "resnetModelHAM1000M.h5"
test_image_path = r"C:\Users\muham\Downloads\HAM10000\HAM1000_image\ISIC_0027753.jpg"

# Load models
model1 = load_model(model1_path)
model2 = load_model(model2_path)
model3 = load_model(model3_path)
model4 = load_model(model4_path)

# Define input sizes for each model
input_size1 = (58, 58, 3)   # For v3p2.h5
input_size2 = (224, 224, 3) # For VGG16, DenseNet, and ResNet

# Preprocess function 1 (for model1)
# def preprocess_image1(image_path, target_size):
#     img = load_img(image_path, target_size=target_size)  # Resize the image




    
#     img_array = img_to_array(img)  # Convert to array
#     # img_array = img_array / 255.0  # Normalize if required
#     return np.expand_dims(img_array, axis=0)  # Add batch dimension

# Preprocess function 1 (for model1)
def preprocess_image1(image_path, target_size):
    # Read the image using OpenCV
    img = cv.imread(image_path)  
    
    # Resize the image to the target size using INTER_AREA interpolation
    img = cv.resize(img, target_size, interpolation=cv.INTER_AREA)
    
    # Convert the resized image to an array (H, W, C)
    img_array = img_to_array(img)  
    
    # Expand dimensions to add batch size (1, H, W, C)
    img_array = np.expand_dims(img_array, axis=0)  
    
    # Return the preprocessed image
    return img_array



# Preprocess function 2 (for model2, model3, model4)
def preprocess_image2(image_path, target_size):
    image = Image.open(image_path).resize(target_size, Image.Resampling.LANCZOS)
    image_array = np.asarray(image)  # Convert to numpy array
    image_array = np.expand_dims(image_array, axis=0)  # Add batch dimension
    return image_array  # No normalization applied

# Preprocess images for each model
image1 = preprocess_image2(test_image_path, target_size=input_size1[:2])  # Compatible with model1
image2 = preprocess_image1(test_image_path, target_size=input_size2[:2])  # Compatible with model2
image3 = preprocess_image1(test_image_path, target_size=input_size2[:2])  # Compatible with model3
image4 = preprocess_image1(test_image_path, target_size=input_size2[:2])  # Compatible with model4

# Get predictions from all models
pred1 = model1.predict(image1)
pred2 = model2.predict(image2)
pred3 = model3.predict(image3)
pred4 = model4.predict(image4)

# Ensemble the predictions (average)
ensemble_pred = (pred1 + pred2 + pred3 + pred4) / 4

# Map predictions to class labels
label_mapping = {
    0: 'nv',
    1: 'mel',
    2: 'bkl',
    3: 'bcc',
    4: 'akiec',
    5: 'vasc',
    6: 'df'
}

# Print individual model predictions
print(f"VGG16 Prediction: {label_mapping[np.argmax(pred2, axis=1)[0]]}")
print(f"Our Own Model Prediction: {label_mapping[np.argmax(pred1, axis=1)[0]]}")
print(f"DenseNet Prediction: {label_mapping[np.argmax(pred3, axis=1)[0]]}")
print(f"ResNet Prediction: {label_mapping[np.argmax(pred4, axis=1)[0]]}")

# Print ensembled prediction
final_class = np.argmax(ensemble_pred, axis=1)[0]
print(f"Final Ensembled Prediction: {label_mapping[final_class]}")
```
